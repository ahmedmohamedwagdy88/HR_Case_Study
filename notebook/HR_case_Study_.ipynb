```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
df = df.dropna()
```

```python
df = df.fillna("Unknown")
```

```python
df.duplicated().sum()
```

```python
df['job_title'] = df['job_title'].str.strip()
df['company'] = df['company'].str.strip()
df['city'] = df['city'].str.strip()
```

```python
df['mode'].value_counts()
```

```python
df['linkedin_skills'] = df['linkedin_skills'].str.lower()
```

```python
df['skills_list'] = df['linkedin_skills'].str.split(',')
```

```python
df.shape
```

```python
df['job_title'].value_counts().head(10)
```

```python
df['company'].nunique()
```

```python
df['mode'].value_counts(normalize=True) * 100
```

```python
df['contract'].value_counts(normalize=True) * 100
```

```python
all_skills = df['linkedin_skills'].str.split(',').explode()
all_skills = all_skills.str.strip()
```

```python
all_skills.value_counts().head(10)
```

```python
df['skills_count'] = df['skills_list'].apply(len)
df['skills_count'].mean()
```

```python
df.groupby('sector')['skills_count'].mean().sort_values(ascending=False)
```

```python
df.to_csv("cleaned_hr_case_study.csv", index=False)
```
